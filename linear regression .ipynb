# METHOD OF LEAST SQUARES 

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\anaconda3\headbrain.csv")
data
```

```python
plt.rcParams['figure.figsize'] = (12.0,9.0)
```

```python
x = data['Head Size(cm^3)']
y = data['Brain Weight(grams)']
```

```python
plt.scatter(x,y)
plt.title('HEADBRAIN DATA LINEAR REGRESSION')
plt.xlabel('Head size')
plt.ylabel('Brain Weight')
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)

num=0
den=0
for  i in range(len(x)):
    num += (x[i] - x_mean)*(y[i] - y_mean)
    den += (x[i] - x_mean)**2
m = num / den
c =y_mean - m*x_mean


print(f' M = : {m}')
print(f' C = : {c}')
```

```python
y_pred = m*x+c
plt.scatter(x,y)
plt.plot ([min(x), max(x)], [min(y_pred),max(y_pred)], color='red')
plt.title('HEADBRAIN DATA LINEAR REGRESSION')
plt.xlabel('Head size')
plt.ylabel('Brain Weight')
plt.show()
```
