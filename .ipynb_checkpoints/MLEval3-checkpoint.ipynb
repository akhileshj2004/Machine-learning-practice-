```python
import os
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import numpy as np


class CNNModel:
    def __init__(self, input_shape, num_classes):
        self.model = self.build_model(input_shape, num_classes)

    def build_model(self, input_shape, num_classes):
        model = Sequential([
            Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=input_shape),
            MaxPooling2D((2, 2)),
            Dropout(0.25),
            
            Conv2D(64, (3, 3), activation='relu', padding='same'),
            MaxPooling2D((2, 2)),
            Dropout(0.25),
            
            Flatten(),
            Dense(128, activation='relu'),
            Dropout(0.5),
            Dense(num_classes, activation='softmax')
        ])
        return model

    def compile(self):
        self.model.compile(optimizer='adam',
                           loss='categorical_crossentropy',
                           metrics=['accuracy'])

    def train(self, train_data, val_data, epochs=20):
        early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
        history = self.model.fit(
            train_data,
            validation_data=val_data,
            epochs=epochs,
            callbacks=[early_stopping]
        )
        return history

    def evaluate(self, test_data):
        return self.model.evaluate(test_data)

    def save(self, file_path):
        self.model.save(file_path)


def load_dataset(train_dir, test_dir, target_size=(32, 32), batch_size=64):
    datagen_train_val = ImageDataGenerator(rescale=1.0 / 255.0, validation_split=0.2)
    datagen_test = ImageDataGenerator(rescale=1.0 / 255.0)
    
    train_generator = datagen_train_val.flow_from_directory(
        train_dir,
        target_size=target_size,
        batch_size=batch_size,
        class_mode='categorical',
        subset='training'
    )
    
    val_generator = datagen_train_val.flow_from_directory(
        train_dir,
        target_size=target_size,
        batch_size=batch_size,
        class_mode='categorical',
        subset='validation'
    )
    
    test_generator = datagen_test.flow_from_directory(
        test_dir,
        target_size=target_size,
        batch_size=batch_size,
        class_mode='categorical',
        shuffle=False
    )
    
    return train_generator, val_generator, test_generator


# Ask the user for the dataset paths
train_path = r"C:\Users\Administrator\Downloads\nature\seg_train\seg_train"
test_path = r"C:\Users\Administrator\Downloads\nature\seg_test\seg_test"

# Load custom dataset
train_data, val_data, test_data = load_dataset(train_path, test_path)

# Get input shape and number of classes
input_shape = train_data.image_shape
num_classes = len(train_data.class_indices)

# Instantiate and use the CNNModel class
cnn = CNNModel(input_shape=input_shape, num_classes=num_classes)
cnn.compile()

# Train the model with early stopping
history = cnn.train(train_data, val_data, epochs=20)

# Plot Loss vs. Epochs
def plot_loss_vs_epochs(history):
    plt.figure(figsize=(8, 5))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Loss vs. Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid()
    plt.show()

plot_loss_vs_epochs(history)

# Evaluate the model on the test set
test_labels = test_data.classes  # Get true labels
test_predictions = np.argmax(cnn.model.predict(test_data), axis=1)  # Predicted classes

# Confusion Matrix
def plot_confusion_matrix(true_labels, predicted_labels, class_names):
    cm = confusion_matrix(true_labels, predicted_labels)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
    disp.plot(cmap=plt.cm.Blues, xticks_rotation='vertical')
    plt.title("Confusion Matrix")
    plt.show()

class_names = list(train_data.class_indices.keys())
plot_confusion_matrix(test_labels, test_predictions, class_names)
```

```python
# Evaluate the model on the test set
test_loss, test_accuracy = cnn.evaluate(test_data)

# Print the testing accuracy
print(f"Testing Accuracy: {test_accuracy * 100:.2f}%")
```
