# BATCH GRADIENT DESCENT

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\anaconda3\example_data.csv")
data
```

```python
plt.rcParams['figure.figsize'] = (12.0,9.0)
```

```python
x = data.iloc[:,0]
y = data.iloc[:,1]
plt.scatter(x,y)
plt.title('data')
```

```python
###  APPLY NORMALISATION IF NEEDED 
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

#Assuming 'data' is your DataFrame
#Replace column indices (0 and 1) with actual column names or use iloc directly

#List of column indices you want to normalize
columns_to_normalize = [0, 1]

#Initialize MinMaxScaler
scaler = MinMaxScaler()

#Normalize each column by index
for col_index in columns_to_normalize:
    Normal = data.iloc[:, col_index].values.reshape(-1, 1)
    data.iloc[:, col_index] = scaler.fit_transform(Normal)
    
    # Print the max and min values after normalization
    print(f"The Max Value of column {col_index} is: {data.iloc[:, col_index].max()}")
    print(f"The Min Value of column {col_index} is: {data.iloc[:, col_index].min()}")
    
    print(f"Normalized - column {col_index}")
    print(data.iloc[:, col_index])
```

## LEARNINIG RATE 0.0001

```python
from sklearn.metrics import mean_squared_error
# Initialize variables
m = 0  # initial slope
c = 0  # initial intercept
l = 0.0001  # learning rate
epochs = 2000  # number of iterations
n = float(len(x))  # number of elements in x

# Lists to store values for plotting
am_values = []
ac_values = []
amse_values = []

# Variables to track the minimum MSE and corresponding epoch, m, and c values
min_mse = float('inf')
best_epoch = 0

best_m = 0
best_c = 0

# Perform gradient descent
for i in range(epochs):
    y_pred = m * x + c  # predicted y values
    d_m = (-1/n) * sum(x * (y - y_pred))  # derivative wrt m
    d_c = (-1/n) * sum(y - y_pred)  # derivative wrt c
    m = m - l * d_m  # update m
    c = c - l * d_c  # update c
    
    mse = mean_squared_error(y, y_pred)
    
    
    # Store the values for plotting
    am_values.append(m)
    ac_values.append(c)
    amse_values.append(mse)

    
    print('epoch= ',i)
    print(f'M = {m}')
    print(f'C = {c}')
    print(f'Mean Squared Error: {mse}')
```

```python
# Plotting Mean Squared Error over epochs
plt.figure(figsize=(10, 6))
plt.plot(amse_values, label="Mean Squared Error")
plt.xlabel("Epochs")
plt.ylabel("Mean Squared Error")
plt.title("MSE vs Epochs for Gradient Descent")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Check if the current mse is the smallest one so far
if mse < min_mse:
        min_mse = mse
        best_epoch = i
        best_m = m
        best_c = c
        # Print the results
        print(f'The epoch with the least MSE is: {best_epoch}')
        print(f'Best M = {best_m}')
        print(f'Best C = {best_c}')
        print(f'Least Mean Squared Error: {min_mse}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

# Learning rates to test
learning_rates = [0.001, 0.0005, 0.0001]

# Number of iterations
epochs = 2000

# Plotting MSE for different learning rates
plt.figure(figsize=(12, 8))

for lr in learning_rates:
    # Initialize variables
    m = 0
    c = 0
    n = float(len(x))
    
    # List to store MSE values
    mse_values = []
    
    # Perform gradient descent
    for i in range(epochs):
        y_pred = m * x + c
        d_m = (-1/n) * sum(x * (y - y_pred))
        d_c = (-1/n) * sum(y - y_pred)
        m = m - lr * d_m
        c = c - lr * d_c
        
        mse = mean_squared_error(y, y_pred)
        mse_values.append(mse)
    
    # Plot the MSE for the current learning rate
    plt.plot(mse_values, label=f"Learning Rate = {lr}")

# Add labels and title
plt.xlabel("Epochs")
plt.ylabel("Mean Squared Error")
plt.title("MSE vs Epochs for Different Learning Rates")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# plot for different learning rates in given algos 
plt.plot(amse_values, label="alpha = 0.0001")
plt.plot(bmse_values, label="alpha = 0.0100")
plt.plot(cmse_values, label="alpha = 0.0900")
plt.xlabel("Epochs")
plt.ylabel("Mean Squared Error")
plt.title("MSE vs Epochs for Different Learning Rates")
plt.legend()
plt.grid(True)
plt.show()
```
